This is for all the markers

```python
# Load the loom file
loom_file = 'FetalBrain/Pool_peaks.agg.loom'

with loompy.connect(loom_file) as ds:
    print("Shape of the main matrix (genes x cells):", ds.shape)  # Rows (genes) x Columns (cells)
    print("Available layers:", list(ds.layers.keys()))
    print("Row attributes (gene-level metadata):", list(ds.ra.keys()))
    print("Column attributes (cell-level metadata):", list(ds.ca.keys()))
```

```python
sel = ["Chr", "Start", "End"]
```

```python
#extracting coordinates  of all peaks 
with loompy.connect(loom_file) as ds:
    # Extract row attributes (gene metadata)
    row_attr_df = pd.DataFrame({key: ds.ra[key] for key in sel})
   
# Display first few rows
print("Row Attributes (Genes):\n", row_attr_df.head())
```

```python
row_attr_df.shape
```

```python
row_attr_df.to_csv("FetalBrain/All_peak_FetalBrain.bed",sep="\t")
```

```python
with loompy.connect(loom_file) as ds:
    layer_name = "marker_peaks"  # Replace with the actual layer name
    df = pd.DataFrame(ds.layers[layer_name][:])
    print(df.head())
```

```python
with loompy.connect(loom_file) as ds:
    layer_name = "marker_peaks"  # Replace with the actual layer name
    df = pd.DataFrame(ds.layers[layer_name][:])
    print(df.head())
```

```python
df.shape
```

```python
row_sums = df.sum(axis=1)
```

```python
idx=row_sums[row_sums > 0].index
```

```python
#extracting coordinates  of all peaks 
with loompy.connect(loom_file) as ds:
    # Extract column attributes (cell metadata)
    col_attr_df =pd.DataFrame({key: ds.ca[key] for key in ds.ca.keys()})  # Converts column attributes into a DataFrame
   
# Display first few rows
print("Column Attributes (Cells):\n", col_attr_df.head())
```
