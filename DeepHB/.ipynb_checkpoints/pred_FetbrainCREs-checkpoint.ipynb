## Script for using Crested model for prediciton on marker CREs for selected clusters from Fetl brain data. This is to validate our models on external data.

```python
import os
import sys
import random
import numpy as np
import pandas as pd
import crested
import pickle
```

```python
##local location
dataDir='/home/p541i/MBsnANA/HBana/ISM/Outs/'
out_Dir='/home/p541i/MBsnANA/HBana/ISM/Outs/'
os.chdir(out_Dir)
```

```python
##load bed files, using augmented peaks
bed_files=dataDir+'FetalBrain/markerpeaks/'
chromsizes=dataDir+'extrafiles/chromSizes_canchrom.genome'
adata_ext = crested.import_beds(
    beds_folder=bed_files, chromsizes_file=chromsizes
)  # the regions file is optional for import_beds
adata_ext
```

```python
df=adata_ext.var
df.to_csv("FetalBrain/CREs_sel.txt",sep="\t")
```

```python
with open(os.path.join(out_Dir,'OutfrmGPU/TopicMPCREs_DeepHB_classes.txt'), "r") as file:
    labels = file.read().splitlines()
print(labels)
```

```python
##load module
model_path = dataDir+"extrafiles/deephb_lstmv2_95.keras"

import keras

model = keras.models.load_model(
    model_path, compile=False
)  # 
```

```python
genome_fasta=dataDir+'extrafiles/gencdH38p13r37.fa'
prediction = crested.tl.predict(adata_ext, model,genome=genome_fasta)
```

```python
prediction.shape
```

```python
np.savetxt('FetalBrain/pred_FB_deephb_lstmv2_95.txt', prediction, delimiter='\t', fmt='%.3f')
with open(os.path.join(out_Dir,'FetalBrain/pred_FB_deephb_lstmv2_95.pkl'), 'wb') as f:
       pickle.dump(prediction, f)
exit()
```
